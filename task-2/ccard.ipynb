disdvantage of undersampling refers to loss of valuable data so for this we use oversampling using smot (synthetic minority of sampling technique) it is mostly common used fro oversampling. here we r not geenrating duplicate data points but we r making synthetic data points.

```python
X = data.drop('Class',axis=1)
y = data['Class']
```

```python
X.shape
```

```python
y.shape
```

```python
from imblearn.over_sampling import SMOTE
```

```python
X_res,y_res = SMOTE().fit_resample(X,y)
```

```python
y_res.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_res,y_res,test_size=0.20,random_state=42)
```

```python
log=LogisticRegression()
log.fit(X_train,y_train)
```

```python
y_pred1=log.predict(X_test)
```

```python
accuracy_score(y_test,y_pred1)
```
